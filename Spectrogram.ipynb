```python
import numpy as np
import librosa, librosa.display
import matplotlib.pyplot as plt
from IPython.display import Image

%matplotlib inline 
plt.rcParams['figure.figsize'] = [24, 16]

filename = "./assets/welcomehome_bestrip.wav"
```

```python
data, rate = librosa.load(filename, sr = 48000)
ft = librosa.stft(data, win_length = 512)
ft2 = librosa.power_to_db(np.abs(ft)**2, amin = 1e-9, top_db = 90)
```

```python

librosa.display.specshow(ft2, x_axis ='time', y_axis ='linear',sr = 48000)
plt.ylim([0, 20000])
plt.xlim([0, 25])
```
